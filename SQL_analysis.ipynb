```python
query = """
SELECT
    crash_type,
    COUNT(*) AS occurrence,
    SUM(CASE
        WHEN crash_type = 'INJURY AND / OR TOW DUE TO CRASH'
        THEN 1
        ELSE 0
    END) AS crash_count,
    (SUM(CASE
        WHEN crash_type = 'INJURY AND / OR TOW DUE TO CRASH'
        THEN 1
        ELSE 0
    END) * 100.0 / COUNT(*)) AS crash_percentage
FROM
    traffic_crashesdata
GROUP BY
    crash_type
ORDER BY
    crash_percentage;
"""

df12 = pd.read_sql(query, engine)
df12
```

```python
query = """
SELECT
    CRASH_DAY_OF_WEEK,
    AVG(crashes_per_hour) AS AVG_CRASHES_PER_HOUR
FROM (
    SELECT
        CRASH_DAY_OF_WEEK,
        CRASH_HOUR,
        COUNT(*) AS crashes_per_hour
    FROM TRAFFIC_CRASHESDATA
    GROUP BY CRASH_DAY_OF_WEEK, CRASH_HOUR
) AS hourly_crashes
GROUP BY
    CRASH_DAY_OF_WEEK
ORDER BY
    AVG_CRASHES_PER_HOUR DESC
"""


#outer query -  Which day has the highest average crashes per hour?
#inner query -  How many crashes happened in each hour of each day?
df20 = pd.read_sql(query,engine)
df20
```

#### __Identify high-risk time slots:__
##### ● Group hours into buckets (Morning, Afternoon, Evening, Night)
##### ● Find which bucket has the highest injury crashes

```python
## General Understnading 
# - morning 5 am to 11 am
# - Afternoon 12 pm to 3 pm
# - Evening 4 pm to 7 pm
# - Night 8 pm to 4 am
```

```python
query = """
SELECT
    TIME_FRAME_BUCKET,
    COUNT(*) AS INJURY_CRASH_COUNT
FROM (
    SELECT
        CASE
            WHEN CRASH_HOUR BETWEEN 5 AND 11 THEN 'Morning'
            WHEN CRASH_HOUR BETWEEN 12 AND 15 THEN 'Afternoon'
            WHEN CRASH_HOUR BETWEEN 16 AND 21 THEN 'Evening'
            ELSE 'Night'
        END AS TIME_FRAME_BUCKET
    FROM TRAFFIC_CRASHESDATA
    WHERE INJURIES_TOTAL > 0
) AS time_bucketed_crashes
GROUP BY
    TIME_FRAME_BUCKET
ORDER BY
    INJURY_CRASH_COUNT DESC;
"""

df21 = pd.read_sql(query,engine)
df21
```

#### __Find the top 3 contributing causes for each crash type.__
##### ○  (Use window functions like ROW_NUMBER() or RANK())

```python
query = """
SELECT
    CRASH_TYPE,
    PRIM_CONTRIBUTORY_CAUSE,
    TOTAL_CRASHES
FROM (
    SELECT
        CRASH_TYPE,
        PRIM_CONTRIBUTORY_CAUSE,
        COUNT(*) AS TOTAL_CRASHES,
        ROW_NUMBER() OVER (
            PARTITION BY CRASH_TYPE
            ORDER BY COUNT(*) DESC
        ) AS rn
    FROM TRAFFIC_CRASHESDATA
    GROUP BY
        CRASH_TYPE,
        PRIM_CONTRIBUTORY_CAUSE
) AS ranked_causes
WHERE rn <= 3
ORDER BY
    CRASH_TYPE,
    TOTAL_CRASHES DESC;
"""

df22 = pd.read_sql(query, engine)
df22
```

#### __Calculate the year-over-year growth rate of crashes.__ 
##### ○ (Use LAG() window function)

```python
### First group by years and total_crash
### Then using LAG window function get the previous year , ordering by years as prev_year_crash
### Now yeay over year growth = (current year - previous year)*100/previous year
```

```python
query = """
SELECT 
    year,
    total_crashes,
    LAG(total_crashes) OVER (ORDER BY year) AS prev_year_crashes,
    ROUND(
        (total_crashes - LAG(total_crashes) OVER (ORDER BY year)) * 100.0
        / LAG(total_crashes) OVER (ORDER BY year),
        2
    ) AS YEAR_GROWTH_RATE
FROM (
    SELECT 
        year,
        COUNT(*) AS total_crashes
    FROM traffic_crashesdata
    GROUP BY year
) AS yearly_data
ORDER BY year
"""

df24 = pd.read_sql(query, engine)
df24
```
